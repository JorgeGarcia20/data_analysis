```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
y = x ** 2
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'y')
plt.subplot(1, 2, 2)
plt.scatter(x, y)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'y')
plt.plot(y, x, 'r')
plt.subplot(1, 2, 2)
plt.bar(x, y)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'y')
plt.bar(y, x)
plt.subplot(1, 2, 2)
plt.bar(x, y)
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(y, x, 'rs-')
plt.subplot(2, 1, 2)
plt.plot(x, y, 'o')
plt.show()
```
