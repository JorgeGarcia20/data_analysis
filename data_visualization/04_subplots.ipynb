```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'y')
plt.show()
```

```python
fig, axes = plt.subplots(2, 4)
axes[0, 0].plot(x, np.sin(y), 'b')
axes[0, 1].plot(x, np.cos(y), 'r')
axes[0, 2].plot(x, np.tan(y), 'g')
axes[0, 3].plot(x, np.arctan(y), 'y')
axes[1, 0].plot(x, np.arcsin(y), 'r')
axes[1, 1].plot(x, np.arccos(y), 'y')
axes[1, 2].plot(x, np.arcsinh(y), 'r')
axes[1, 3].plot(x, np.arctan(y), 'g')
fig.tight_layout()

plt.show()
```

# Contexto a las gráficas

```python
z = np.cos(x)
```

```python
plt.style.use('dark_background')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8,5))

axes[0].plot(x, y, 'y', label='$sin(x)$')
axes[0].set_title('Relacion x - y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()


axes[1].plot(x, z, 'g', label='$cos(x)$')
axes[1].set_title('Relacion x - z')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Z')
axes[1].legend()

fig.tight_layout()

plt.show()
```

## Estilos matlab

```python
fig, ax = plt.subplots(figsize=(6,6))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'y-')
ax.plot(x, x+3, 'b.-')
ax.plot(x, x+4, 'go:')

plt.show()
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(6,6))

ax.plot(x, x+1, color='red', alpha=0.8, linewidth=4, linestyle='-', marker='o', markersize=10, markerfacecolor='#ffffff')
ax.plot(x, x+2, color='yellow', linewidth=2, linestyle='--', marker='x', markersize=12, markerfacecolor='#cccccc')
ax.plot(x, x+3, color='blue', linewidth=2, linestyle=':', marker='p', markersize=10)
ax.plot(x, x+4, color='#ffffff')

plt.show()
```
