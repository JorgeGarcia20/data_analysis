```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 10)
y = x**2 
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.8])
axes.plot(x, y, 'b--')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes_2 = fig.add_axes([0.2, 0.6, 0.4, 0.3])
axes_2.set_facecolor('gray')

axes.plot(x, y, 'b')
axes_2.plot(y, x, 'r')

plt.show()
```
