```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'BRAZIL', 'COLOMBIA']
populations = [10000, 4222, 3240, 2444, 2823]
```

```python
plt.bar(country, populations, width=0.5, color=['red', 'green', 'blue', 'yellow', 'pink'])
# plt.title('Poblacion de los paises')
plt.xlabel('Países')
plt.ylabel('Población')
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Brazil', 'Colombia'), rotation=45)
```

```python
plt.barh(country, populations, color=['red', 'green', 'blue', 'yellow', 'pink'])
plt.show()
```

# Otros graficos

```python
data = np.random.randint(1, 50, 100)
```

## Histograma

```python
plt.hist(data, bins=20)
plt.show()
```

## Boxplot

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

## Dispersion o scatter


```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
plt.scatter(x, y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
