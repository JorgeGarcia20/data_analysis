```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.barplot(x=['A', 'B', 'C'], y=[4, 2, 5])
plt.show()
```

```python
sns.set(style='dark', palette='dark', font="JetBrains Mono", font_scale=1)
sns.barplot(x=['A', 'B', 'C'], y=[4, 2, 5])
plt.show()
```

## Parámetros más usados de `seaborn`

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, 
            x='total_bill', 
            hue='sex', 
            kind='kde',
            legend=True, 
            palette='dark', 
            alpha=0.25)
plt.show()
```
