```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.histplot(data=tips, 
             x='tip', 
             bins=12, 
             cumulative=False, 
             hue='sex', 
             stat='count', 
             multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', fill=True, bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='smoker', stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='kde', multiple='stack')
plt.show()
```

## Graficas categoricas

```python
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
sns.countplot(data=tips, y='day', hue='sex')
plt.show()
```

```python
sns.stripplot(data=tips, 
              x='day', 
              y='total_bill',
              hue='sex', 
              dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, 
              x='day', 
              y='total_bill',
              hue='sex', 
              dodge=True)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, 
              x='day', 
              y='total_bill',
              hue='sex', 
              dodge=True)

sns.swarmplot(data=tips, 
              x='day', 
              y='total_bill',
              hue='sex', 
              dodge=True, color='0', marker='p')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(data=tips, 
              x='day', 
              y='total_bill',
              hue='sex', 
               split=True,
              dodge=True,
              palette='pastel')
plt.show()
```

```python
sns.violinplot(data=tips, y='total_bill')
plt.show()
```

```python
sns.catplot(data=tips, 
            x='day', 
            y='total_bill', 
            hue='sex', 
            dodge=True, 
            kind='box', 
            col='time')
plt.show()
```

## Graficos de tipo relación

```python
plt.figure(figsize=(8,8))
markers = {'Lunch': 'D', 'Dinner':'s'}
sns.scatterplot(data=tips, 
                x='total_bill', 
                y='tip', 
                hue='day', 
                style='time',
                size='size',markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
plt.figure(figsize=(8,8))
markers = {'Lunch': 'D', 'Dinner':'s'}
sns.lineplot(data=tips, 
                x='total_bill', 
                y='tip', 
                hue='day', 
                style='time',
                size='size',
             markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.relplot(data=tips, 
            x='total_bill', 
            y='tip', 
            hue='day', 
            style='time',
            size='size',
            col='time')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
sns.pairplot(data=tips, hue='day')
plt.show()
```

## Joinplot y Pairplot

```python
sns.jointplot(data=tips, 
              x='total_bill', 
              y='tip', 
              hue='sex', 
              kind='hist', 
              marginal_ticks=dict(bins=25, 
                                  fill=False, 
                                  multiple='dodge'))
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True)
plt.show()
```
